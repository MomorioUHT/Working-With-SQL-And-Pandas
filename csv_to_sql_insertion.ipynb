### Imports

```python
import os
from dotenv import load_dotenv
import mysql.connector
import pandas as pd
from sqlalchemy import create_engine, URL
```

```python
load_dotenv()

host = os.getenv("MYSQL_HOST")
port = os.getenv("MYSQL_PORT")
user = os.getenv("MYSQL_USER")
password = os.getenv("MYSQL_PWD")
database = os.getenv("MYSQL_DATABASE")
```

### Connect and get cursor

```python
pool = mysql.connector.connect(
    host=host,
    port=port,
    user=user,
    password=password,
    database=database
)

cursor = pool.cursor()
```

### Read data from CSV

```python
df = pd.read_csv("datasets/Mobile_Price_2025.csv")
df.head(3)
```

```python
df.info()
```

### use connection_url as an object (safer when password contains illegal char)

```python
connection_url = URL.create(
    drivername="mysql+pymysql",
    username=user,
    password=password,
    host=host,
    port=port,
    database=database,
)

engine = create_engine(connection_url)
```

### Automatic table creation

```python
import cryptography

df.to_sql(
    name="smartphones",
    con=engine,
    if_exists="replace",
    index=False
)
```
